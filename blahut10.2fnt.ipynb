## Polynomials

```python
n = 5
p = 31
```

```python
divmod((p-1), n)
```

```python
d = Matrix([2, 6, 4, 4, 0])
g = Matrix([3, 0, 2, 1, 2])
```

```python
plot_pdf(doc[328], [.60, .95])
```

```python
dk = Matrix([[divmod(2**(i*k), p)[1] for i in range(n)] for k in range(n) ])
dk
```

```python
dd = dk*d
dd
```

```python
dd1 = Matrix([divmod(i, p)[1] for i in dd.flat()])
dd1
```

```python
gg = dk*g
gg
```

```python
gg1 = Matrix([divmod(i, p)[1] for i in gg.flat()])
gg1
```

```python
plot_pdf(doc[329], [0, .435])
```

```python
ss = sy.hadamard_product(dd1, gg1)
ss
```

```python
ss1 = Matrix([divmod(i, p)[1] for i in ss.flat()])
ss1
```
