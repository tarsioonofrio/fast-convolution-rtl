```python
mk = [3, 4, 5]
sy.Matrix(mk)
```

```python
3*5*17
```

```python
x = 17
x
```

```python
ck = [divmod(x, m)[1] for m in mk]
ck
```

```python
mm = np.prod(mk)
mm
```

```python
mmk = sy.Matrix([sy.expand(np.prod(d)) for d in itertools.combinations(reversed(mk), len(mk)-1)])
mmk
```

```python
gcd = [sy.gcdex(mm, m) for mm, m in zip(mmk, mk)]
sy.Matrix(gcd)
```

```python
nnk = [g[0] for g in gcd]
sy.Matrix(nnk)
```

```python
nk = [g[1] for g in gcd]
sy.Matrix(nk)
```

```python
cx_sum = sum([c*m*g for c, m, g in zip(ck, mmk, nnk)])
cx_sum
```

```python
cx = divmod(int(cx_sum), mm)[1]
cx
```
