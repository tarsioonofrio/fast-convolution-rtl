```python
def multiplication_trick2(x0, x1, y0, y1):
    ax0 = x0+x1
    ax1 = x0-x1
    bx0 = (y0+y1)/2
    bx1 = (y0-y1)/2

    m0 = ax0*bx0
    m1 = ax1*bx1
    s0 = m0 + m1
    s1 = m0 - m1
    return (s0, s1, m1, m0)
```
