# S2 livro

```python
ma = sy.Matrix([[1, 1], [1, 0], [0, 1]])
ma
```

```python
md = sy.Matrix(di)
md
```

```python
ma * md
```

```python
mb = sy.Matrix([[1, 0], [-1, 1], [1, 1]])
mb
```

```python
mg = sy.Matrix(gi)
mg
```

```python
mbg = sy.diag(*(mb * mg).tolist())
mbg
```

```python
s2b = sy.diag(gi[0], gi[1] - gi[0], gi[0] + gi[1])
s2b
```

```python
mc = sy.Matrix([
    [1, 0, -1],
    [1, 1, 0],
])

mc
```

```python
sy.MatMul(mc, mbg, ma, md)
```

```python
sy.expand(mc * mbg * ma * md)
```
