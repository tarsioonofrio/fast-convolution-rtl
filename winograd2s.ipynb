```python
di = sy.symbols(" ".join(f"d{i}^(2)"for i in range(g_num)))
di
```

```python
gi = sy.symbols(" ".join(f"g{i}^(2)"for i in range(g_num)))
gi
```

```python
# S2 livro
```

```python
a_mtx = sy.Matrix([[1, 1], [1, 0], [0, 1]])
a_mtx
```

```python
d_mtx = sy.Matrix(di)
d_mtx
```

```python
a_mtx * d_mtx
```

```python
b_mtx = sy.Matrix([[1, 0], [-1, 1], [1, 1]])
b_mtx
```

```python
g_mtx = sy.Matrix(gi)
g_mtx
```

```python
bg_mtx = sy.diag(*(b_mtx * g_mtx).tolist())
bg_mtx
```

```python
c_mtx = sy.Matrix([
    [1, 0, -1],
    [1, 1, 0],
])

c_mtx
```

```python
sy.MatMul(c_mtx, bg_mtx, a_mtx, d_mtx)
```

```python
sy.expand(c_mtx * bg_mtx * a_mtx * d_mtx)
```
